### **Outlier Values**

Outlier adalah nilai yang sangat jauh menyimpang dari nilai data lainnya. Dapat di periksa menggunakan boxplot pada kolom-kolom numerik yang sudah dibersihkan dari ERROR/UNKNOWN.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers Values**

Outlier adalah nilai yang sangat jauh menyimpang dari nilai data lainnya, Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Outlier pada kolom Total Spent tidak perlu ditangani karena nilai tersebut masih masuk akal. Nilai tinggi pada Total Spent bisa terjadi karena pelanggan membeli item dalam jumlah banyak dengan harga yang tinggi, yang memang bisa saja terjadi. Jadi outlier di sini bukan merupakan kesalahan data, melainkan memang kondisi transaksi yang sebenarnya terjadi sehingga tidak perlu diubah atau dihapus.

## **ScatterPlot**

## **Bubble Chart**

**Perbandingan Produk: Harga Satuan, Rata-rata Quantity, dan Total Pendapatan**

Visualisasi ini menampilkan hubungan antara jumlah produk yang dibeli (Quantity), total pengeluaran pelanggan (Total Spent), dan jenis produk (Item), sehingga dapat membantu melihat produk mana yang cenderung menghasilkan transaksi dengan nilai lebih tinggi.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Quantity', y='Total Spent', size='Price Per Unit', hue='Item', data=df, sizes=(50, 1000), alpha=0.7, palette='viridis')
plt.title('Hubungan Quantity dan Total Spent (Ukuran: Harga Produk, Warna: Item)', fontsize=16)
plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Total Spent', fontsize=12)
plt.grid(True)
plt.legend(title='Item', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Dari visualisasi ini dapat terlihat bahwa transaksi dengan quantity yang lebih tinggi cenderung menghasilkan total pembayaran yang lebih besar. Selain itu, gelembung yang lebih besar menunjukkan produk dengan harga satuan yang lebih tinggi. Hal ini berarti meskipun jumlah produk yang dibeli tidak terlalu banyak, total pembayaran tetap bisa tinggi jika produk tersebut memiliki harga yang lebih mahal.

Berdasarkan hasil tersebut, kafe dapat meningkatkan pendapatan dengan mendorong pelanggan untuk membeli lebih banyak produk dalam satu transaksi, misalnya melalui paket menu atau promo bundling. Selain itu, produk dengan harga lebih tinggi yang tetap diminati pelanggan dapat dipromosikan sebagai menu rekomendasi agar lebih sering dibeli.

## **Heatmap**

**Heatmap Korelasi Antar Variabel Numerik**

Melihat hubungan antar variabel numerik yaitu, Quantity, Price Per Unit, dan Total Spent.

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Quantity','Price Per Unit','Total Spent']].corr(),
            annot=True,
            cmap='coolwarm')
plt.title("Heatmap Korelasi Variabel")
plt.show()
```

Dari hasil heatmap terlihat bahwa total spent memiliki hubungan yang cukup kuat dengan quantity dan juga price per unit. Hal ini menunjukkan bahwa dua faktor utama yang mempengaruhi total pengeluaran pelanggan adalah jumlah produk yang dibeli dan harga produk tersebut. Jika pelanggan membeli lebih banyak item atau jika mereka memilih produk dengan harga yang lebih tinggi, maka total pembayaran mereka juga akan meningkat.

Bedasarkan hal tersebut, hal yang dapat dilakukan untuk memaksimalkan pendapatkan adalah dengan menonjolkan produk dengan harga yang sedikit lebih tinggi tetapi tetap memiliki kualitas yang baik dan diminati pelanggan. Selain itu juga dapat mendorong pelanggan untuk membeli lebih dari satu item dalam satu transaksi, bisa dengan menawarkan melalui kasir.
